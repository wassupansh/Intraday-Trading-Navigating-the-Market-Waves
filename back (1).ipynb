```python
data1 = data.iloc[::,0:8]
data1 = data1.drop(["Date","olhc4"],axis=1)
data1["macd_y"] = data["macd_y"]
data1["supertrend_y"] = data["supertrend_y"]
data1["rsi_y"] = data["rsi_y"]
data1["dema_y1"] = data["dema_y1"]
data1["tema_y"] = data["tema_y"]
data1["adx_y1"] = data["adx_y1"]
data1["adx_y2"] = data["adx_y2"]
data1["adx_di_y"] = data["adx_di_y"]
data1["ha_signals_y"] = data["ha_signals_y"]
data1["williams_y"] = data["williams_y"]
#data1["ha_signals_y111"] = data["ha_signals_y111"]
#data1["col_y"] = data["col_y"]
```

```python
import pandas as pd
from backtesting import Backtest, Strategy

class Signal(Strategy):
    
    def init(self):
        pass

    def next(self):
        indicator_value = self.data['williams_y']
        if indicator_value == 1:
            if not self.position:
                self.buy()
        elif indicator_value == 0:
            if self.position:
                self.position.close()
        elif indicator_value == -1:
            if not self.position:
                self.sell()
#sl 
# Assuming your data is in a DataFrame called 'data1' with a column named 'macd_y'
# Convert the index to datetime if it's not already
data1.index = pd.to_datetime(data1.index)

bt = Backtest(data1,Signal, cash=10_000)
stats = bt.run()
print(stats)
```
